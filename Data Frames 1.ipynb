# astype() method    # no null values

```python
nba = pd.read_csv("nba.csv").dropna(how="all") # no NaN
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None", inplace=True)
nba.head(6)
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba.head(3)
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Weight"] = nba["Weight"].astype("int")
nba.head(3)
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category") #reduces memory usage by adding a category when few datasets in a column are the same
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")
```

# rank() method

```python
nba = pd.read_csv("nba.csv").dropna(how="all") # no NaN allowed and no floats
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
nba.head(7)
```

```python
nba.sort_values(by="Salary Rank", ascending=True).head(7)
```
