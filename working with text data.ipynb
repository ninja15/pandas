# lower() upper() title() len() methods

```python
chicago = pd.read_csv("chicago.csv")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

```python
chicago["Name"] = chicago["Name"].str.title()
chicago["Position Title"] = chicago["Position Title"].str.title()
chicago.head(3)
```

```python
chicago["Department"].str.len().head(3)
```

# .str.replace() method

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT", "MANAGEMENT")
chicago.head()
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$", "").astype(float)
chicago.head()
```

# filtering with string methods

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask].head()
```

```python
mask = chicago["Position Title"].str.lower().str.startswith("water")
chicago[mask].head()
```

```python
chicago[chicago["Position Title"].str.lower().str.endswith("ist")].head()
```

# strip() lstrip() rstrip() methods

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
# lstrip() left white space remove, rstrip() removes the right space, strip() removes left and right
```

```python
chicago["Name"] = chicago["Name"].str.strip()
chicago.head(3)
```

# string methods on index and columns

```python
chicago = pd.read_csv("chicago.csv",index_col="Name").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago.index = chicago.index.str.strip().str.title()
chicago.head()
```

```python
chicago.columns = chicago.columns.str.upper()
chicago.head(3)
```

# split strings by characters with .str.split() method

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago["Name"].str.split(",").str.get(0).str.title().value_counts() # str.get() with argument of position within a list 
```

```python
chicago["Position Title"].str.split(" ").str.get(0).value_counts().head()
```

```python
chicago["Name"].str.split(",").str.get(1).str.lstrip().str.split(" ").str.get(0).value_counts().head()
```

# expand and n parameters with .str.split() method

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",", expand=True)
chicago.head()
```

```python
chicago["Position Title"].str.split(" ", expand=True).head()
```

```python
chicago[["First", "Remainders"]] = chicago["Position Title"].str.split(" ", expand=True, n=1)
chicago.head()
```
