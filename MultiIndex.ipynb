# extract rows from a MultiIndex dataframe

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True) # faster when sorted
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01", "Brazil"), "Price in US Dollars"]
```

```python
bigmac.loc[("2015-07-01", "Brazil"), "Price in US Dollars"]
```

```python
bigmac.ix[("2016-01-01", "Russia"), "Price in US Dollars"]
```

```python
bigmac.loc[("2016-01-01", "Brazil") : ("2016-01-01", "Estonia")] # extracting multiple rows 
```

# the .transpose() method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True) # faster when sorted
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
bigmac.head(1)
```

```python
bigmac.ix["Price in US Dollars", ("2016-01-01", "Denmark")]
```

# .unstack() method

```python
world = pd.read_csv("worldstats.csv", index_col=["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack().unstack().unstack()
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack(1)
```

```python
s.unstack("country")
```

```python
world = pd.read_csv("worldstats.csv", index_col=["country", "year"])
s= world.stack()
s.head(3)
```

```python
s.unstack(level=[1,0])
```

```python
s.unstack(level=["country","year"])
```

```python
s = s.unstack("year", fill_value=0)
s.head(3)
```

# .pivot() method

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales = sales.pivot(index="Date", columns="Salesman", values="Revenue")
sales.head(3)
```

# pivot_table() method

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], columns="City", aggfunc="sum")
```
