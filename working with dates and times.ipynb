# create range o dates with pd.date_range() method

```python
pd.date_range(start="2016-01-01", end="2016-01-10", freq="D")
```

```python
times = pd.date_range(start="2016-01-01", end="2016-01-10", freq="D") # at least 2 parameters required
times[0]
```

```python
pd.date_range(start="2016-01-01", end="2016-01-10", freq="H") 
#freq="D" means days, 
#freq="B" means bussines days  
#freq="W" means week starting with sunday "W-FRI" starts with friday
#freq="H" starts with hour, "6H" six hours periods
```

```python
pd.date_range(start="2016-01-01", end="2016-12-10", freq="MS") 
#freq="M" means month ends "MS" month start
```

```python
pd.date_range(start="2016-01-01", end="2050-12-10", freq="AS") 
#freq="A" means year ends "AS" year start
```

```python
pd.date_range(start="2012-09-09", periods=25, freq="6H") # periods = the numbers of TimeStamps from the start date 
```

```python
pd.date_range(end="1999-12-31", periods=20, freq="W-SUN")
```
