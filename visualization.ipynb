# bar charts

```python
gg = data.DataReader(name="GOOG", data_source="yahoo", start="2004-01-01", end="2016-12-31")
gg.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
gg["Close"].apply(rank_performance).value_counts().plot(kind="bar")
```

```python
gg["Close"].apply(rank_performance).value_counts().plot(kind="barh")
```

# pie charts

```python
apple = data.DataReader(name="AAPL", data_source="yahoo", start="2012-01-01", end="2016-12-31")
apple.head(3)
```

```python
apple["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 93.3617192254726:
        return "Above average"
    else:
        return "Below average"
```

```python
plt.style.use("classic")
apple["Close"].apply(rank_performance).value_counts().plot(kind="pie", legend=True)
```

# histograms

```python
gg = data.DataReader(name="GOOG", data_source="yahoo", start="2004-01-01", end="2016-12-31")
gg.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
gg["High"].apply(custom_round).value_counts().sort_index()
```

```python
gg["High"].apply(custom_round).nunique()
```

```python
gg["High"].apply(custom_round).plot(kind="hist", bins=9)
```
